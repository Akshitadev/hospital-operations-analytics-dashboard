```python
import requests
import pandas as pd

url = "https://jsonplaceholder.typicode.com/users"

response = requests.get(url)

print(response.status_code)

data = response.json()

api_df = pd.DataFrame(data)

api_users = api_df[
    [
        "id",
        "name",
        "username",
        "email",
        "phone",
        "website"
    ]
]

api_users.to_csv("api_users.csv", index=False)
```
