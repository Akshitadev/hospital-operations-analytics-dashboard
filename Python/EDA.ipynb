```python
import numpy as np

def perform_eda(df, table_name, numeric_columns=None, categorical_columns=None):

    print("=" * 70)
    print(f"EDA REPORT : {table_name.upper()}")
    print("=" * 70)

    # Dataset Overview
    print("\n📌 Shape")
    print(df.shape)

    print("\n📌 Columns")
    print(df.columns.tolist())

    print("\n📌 Data Types")
    print(df.dtypes)

    # Data Quality
    print("\n📌 Missing Values")
    print(df.isnull().sum())

    print("\n📌 Duplicate Rows")
    print(df.duplicated().sum())

    # Summary Statistics
    print("\n📌 Summary Statistics")
    print(df.describe(include="all"))

    # NumPy Statistics
    if numeric_columns:

        print("\n📌 NumPy Statistics")

        for column in numeric_columns:

            print(f"\n{column}")

            print("Mean:", np.mean(df[column]))
            print("Median:", np.median(df[column]))
            print("Minimum:", np.min(df[column]))
            print("Maximum:", np.max(df[column]))
            print("Standard Deviation:", np.std(df[column]))

    # Frequency Analysis
    if categorical_columns:

        print("\n📌 Frequency Analysis")

        for column in categorical_columns:

            print(f"\n{column}")

            print(df[column].value_counts())

    print("\n" + "=" * 70)
```

```python
perform_eda(
    doctors_df,
    "Doctors",
    numeric_columns=["experience_years"],
    categorical_columns=["specialization"]
)
perform_eda(
    patients_df,
    "Patients",
    categorical_columns=[
        "gender",
        "blood_group"
    ]
)
perform_eda(
    departments_df,
    "Departments",
    categorical_columns=[
        "department_name",
        "location"
    ]
)

perform_eda(
    appointments_df,
    "Appointments",
    categorical_columns=[
        "appointment_status"
    ]
)

perform_eda(
    encounters_df,
    "Encounters",
    numeric_columns=[
        "consultation_duration"
    ]
)
perform_eda(
    billing_df,
    "Billing",
    numeric_columns=[
        "total_amount"
    ]
)

perform_eda(
    payments_df,
    "Payments",
    numeric_columns=[
        "amount_paid"
    ],
    categorical_columns=[
        "payment_method",
        "payment_status"
    ]
)

perform_eda(
    prescriptions_df,
    "Prescriptions",
    numeric_columns=[
        "duration_days"
    ],
    categorical_columns=[
        "medicine_name"
    ]
)
```

```python
####Patient_Analysis:
import sqlite3

connection = sqlite3.connect(
    "/Users/akshitajoshi/Hospital_Data_Analytics_Internship/Hospital-Operations-Analytics.db"
)
import pandas as pd
import numpy as np

patients_df = pd.read_sql_query(
    "SELECT * FROM Patients;",
    connection
)
appointments_df = pd.read_sql_query("SELECT * FROM appointments;", connection)


print("Total Patients: " , patients_df["patient_id"].nunique())
gender_counts = patients_df["gender"].value_counts()
print(gender_counts)

patient_visits = (
    appointments_df
    .groupby("patient_id")
    .size()
    .reset_index(name="total_appointments")
)
patient_visits



patient_visits.sort_values(
    by="total_appointments",
    ascending=False
).head(10)




top_patients = patient_visits.merge(
    patients_df[
        ["patient_id", "first_name", "last_name"]
    ],
    on="patient_id"
)

top_patients.sort_values(
    by="total_appointments",
    ascending=False
).head(10)




```
